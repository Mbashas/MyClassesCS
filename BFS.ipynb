```python
import networkx as nx
import matplotlib.pyplot as plt

# Create an empty graph
graph = nx.Graph()

# Add nodes to the graph
graph.add_nodes_from([1, 2, 3, 4, 5, 6])

# Add edges to the graph
graph.add_edges_from([(1, 2), (1, 3), (2, 4), (2, 5), (3, 6)])

# Perform BFS traversal
bfs_nodes = list(nx.bfs_tree(graph, 1))

# Create a subgraph with BFS nodes
bfs_graph = graph.subgraph(bfs_nodes)

# Draw the graph
pos = nx.spring_layout(bfs_graph)
nx.draw(bfs_graph, pos, with_labels=True, node_color='lightblue', node_size=500, font_size=12, edge_color='gray')

# Show the graph
plt.show()
```
