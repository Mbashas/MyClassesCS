```python
def newvalue(num_list):
    unique = []
    for num in num_list:
        if num not in unique:
            unique.append(num)
    return unique

inputValue = [1, 2, 3, 2, 4, 5, 1, 3]
outputValue = newvalue(inputValue)
print(outputValue) 
```
